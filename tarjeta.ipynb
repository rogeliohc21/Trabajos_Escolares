```python
class Tarjeta:
  def __init__(self, nombre, numtrjt, banco, tipo) -> None:
      self.nombre = nombre
      self.numtrjt = numtrjt
      self.banco = banco
      self.tipo = tipo

  def __str__(self) -> str:
    return self.nombre + " " + "No de tarjeta: " + " " + str(self.numtrjt) + " " + self.banco + " " + self.tipo


class Credito(Tarjeta):
  def __init__(self, tipo, banco, nombre, numtrjt, saldo) -> None:
     super().__init__(nombre, numtrjt, banco, tipo)
     self.saldo = saldo

class Descuento(Tarjeta):
  def __init__(self, opcion1, opcion2, nombre, tarjetanum, porcentajes, tiendas ) -> None:
    self.opcion1 = opcion1
    self.opcion2 = opcion2
    self.nombre = nombre
    self.tarjetanum = tarjetanum
    self.porcentajes = porcentajes
    self.tiendas = tiendas

def __str__(self) -> str:
    return self.opcion1 + " " + self.opcion2 + " " + self.nombre + " " + "No de tarjeta: " + " " + str(self.tarjetanum) + " " + str(self.porcentajes) + " " + self.tiendas


class Variable(Descuento):
  def __init__(self, nombre, tarjetanum, porcentajes, tiendas, vip) -> None:
      super().__init__(nombre, tarjetanum, porcentajes, tiendas, vip)
      self.vip = vip


class Fijo(Descuento):
  def __init__(self, nombre, tarjetanum, porcentajes, tiendas, regular) -> None:
      super().__init__(nombre, tarjetanum, porcentajes, tiendas, regular)
      self.regular = regular


class Total:
  def __init__(self, monto, pors) -> None:
      self.monto = monto
      self.pors = pors

  def calc_desc(self):
    return self.monto * self.pors / 100


  def __str__(self) -> str:
      return f"el descuesnto  {self.__class__.__name__} es: {self.calc_desc()}"




credito = Credito("Jose Rogelio Hidalgo ", 4436313805855236, "SATURNO BANK", "CREDITO", 5000)
descuento = Descuento("tarjeta descuento fijo", "tarjeta descuento variable", "Jose Rogelio Hidalgo", 1645, 10-15, "SEARS")



print(" BIENVENIDO AL MOODULO DE CONSULTA PARA TARJETABIENTES DE SATURNO BANK")
print("MENU")
print("1) CONSULTA POR NOMBRE")
print("2) CONSULTA POR NUMERO DE TARJETA")
datos=int(input("ESCOGE EL NUMERO DE CONSULTA:  "))
if datos == 1:
  print(credito.nombre)
elif datos == 2:
  print(credito.nombre)
else:
  print("ERROR")
print("-"*50)
print("OPCIONES A ELEGIR POR SER TARJETABIENTE DE SATURNO BANK")
print("1) DESCUENTO FIJO DEL 5% EN TIENDAS SELECCIONADAS")
print("2) DESCUENTO VARIABLE DEL 10 AL 15% EN TIENDAS CON CONVENIO")
opc=int(input("escoge la tarjeta de descuento que gustes:   "))
if opc == 1:
  print(descuento.opcion1)
elif opc == 2:
  print(descuento.opcion2)
  print("-"*50)
  print(descuento.nombre)
  print(descuento.tarjetanum)
  print("OPCIONES DE TIENDAS ")
  print("1) SEARS 12% DE DESUENTO")
  print("2) LIVERPOOL 11% DE DESCUENTO ")
  print("3) PALACIO DE HIERRO 15% DE DESCUENTO")
  print("4) ZARA 10% DE DESCUENTO")
  tienda=int(input("¿QUÉ TIENDA ESCOGES?"))
  if tienda == 1:
    print(descuento.tiendas)
  else:
    print("ERROR")


print("-"*50)
clc=int(input("total de la compra:  "))
res=int(input("descuento otorgado: "))
desc_total=clc*res/100
print("EL DESCUENTO ES DE:", desc_total)
calc=Total(3000, 15)
print(calc.calc_desc())
print(calc)





```
