#**Introducción y justificación**
La contaminación atmosférica es uno de los desafíos ambientales más urgentes de la actualidad, impactando gravemente la salud pública y el equilibrio ecológico. En este contexto, nuestro proyecto se enfoca en abordar este problema desde una perspectiva de sensibilización, reconociendo la importancia de generar conciencia colectiva sobre los efectos de la contaminación y la necesidad de promover prácticas responsables.

Si bien un análisis completo de la huella de carbono es crucial para la sostenibilidad a largo plazo, esta investigación prioriza el diseño de estrategias que maximicen el impacto social mediante una campaña efectiva. En lugar de centrarnos exclusivamente en el impacto directo de los contaminantes, optamos por analizar cómo llegar a un mayor número de personas en un tiempo limitado a través de plataformas digitales. Esto responde a la necesidad de contar con evidencias tangibles y prácticas que respalden el proyecto prototípico y permitan evaluar el alcance de la difusión.

A través de videos promocionales en redes sociales, se busca alcanzar al menos 1500 visitas por plataforma dentro de un presupuesto limitado, con el objetivo de identificar oportunidades de mejora en la comunicación y reforzar la difusión de programas de gestión de la calidad del aire en la Zona Metropolitana del Valle de México. Este enfoque no solo busca reducir la brecha de información entre la población y las iniciativas gubernamentales, sino también potenciar la participación ciudadana en la mejora de la calidad del aire, contribuyendo así a la salud pública y al bienestar colectivo.

#**Planteamiento del problema:**

Objetivo:
Maximizar el alcance de personas (visitas) a través de las plataformas de redes sociales con un presupuesto mínimo, garantizando al menos 1500 visitas por video en cada una de las plataformas.

**Datos del Problema**

1. **Costos promedio por mil impresiones (CPM):**

- Instagram: $222.71 Mxn por cada 1000 impresiones.

- Tiktok: $206.98 Mxn por cada 1000 impresiones.

- Facebook: $202.43 Mxn por cada 1000 impresiones.

- X (estimado): $200 Mxn por cada 1000 impresiones.

2. **Impresiones necesarias para 1500 visitas en cada red social:**

- Instagram: 1500 impresiones / 1000 * 222.71: $334.07 Mxn.

- Tiktok: 1500 impresiones / 1000 x 206.98 : $310.47 Mxn.

- Facebook: 1500 impresiones  1000 * 202.43 : $303.65 Mxn.

- X: 1500 impresiones / 1000 * 200 : $300 Mxn.

3. **Presupuesto total mínimo requerido:
El presupuesto total para garantizar 1500 visitas por plataforma es:**

334.07 + 310.47 + 303.65 + 300 = **1248.19 Mxn.**

#**Modelo de Programación Lineal**

En este modelo, buscamos determinar cuántas impresiones se deben asignar a cada plataforma para maximizar el alcance mientras se asegura de no exceder el presupuesto de $1248.19 MXN.

**Variables de Decisión**

* X1: Número de miles de impresiones en Instagram.

* X2: Número de miles de impresiones en TikTok.

* X3: Número de miles de impresiones en Facebook.

* X4: Número de miles de impresiones en X.

**Función Objetivo**
Maximizar el total de visitas, que es equivalente a maximizar las impresiones en todas las plataformas:

**Maximizar Z = 1000 * ( x1 + x2 + x3 + x4 )**

**Restricciones:**
Garantizar al menos 1500 visitas por plataforma:

1. Instagram (X1) **X1 ≥ 1.5**

2. Tiktok (X2): **X2 ≥ 1.5**

3. Facebook (X3): **X3 ≥ 1.5**

4. X (twiiter) (X4): **X4 ≥ 1.5**

5. Respetar el presupuesto máximo de $1248.19 Mxn.

  **222.71 * X1 + 206.98 * x2 + 202.43 * x3 + 200 * X4 ≤ 1248.19**

6. No puede haber una cantidad negativa de contaminantes: **X1,X2,X3,X4 ≥ 0**

#**Gráfica del área de posibles soluciones**



```python
import numpy as np
import matplotlib.pyplot as plt

# Rango de valores para las variables
x1 = np.linspace(0, 5, 500)  # Hasta 5 mil impresiones como ejemplo

# Restricciones del problema
cost_instagram = 222.71
cost_tiktok = 206.98
cost_facebook = 202.43
cost_x = 200
presupuesto_total = 1248.19

# Restricciones presupuestarias (al resolver por igualdad, despejamos x4)
x4_1 = (presupuesto_total - cost_instagram * x1 - 1.5 * cost_tiktok - 1.5 * cost_facebook) / cost_x  # TikTok y Facebook mínimos
x4_2 = (presupuesto_total - cost_tiktok * x1 - 1.5 * cost_instagram - 1.5 * cost_facebook) / cost_x  # Instagram y Facebook mínimos

# Región factible: considerando restricciones de impresiones mínimas (x4 >= 1.5)
x4_min = 1.5  # Impresión mínima en X
x4_limit = np.minimum(x4_1, x4_2)

# Crear la gráfica
plt.figure(figsize=(10, 8))

# Graficar las restricciones
plt.plot(x1, x4_1, label="Restricción: Instagram y TikTok", color="blue")
plt.plot(x1, x4_2, label="Restricción: TikTok e Instagram", color="orange")
plt.axhline(x4_min, color="red", linestyle="--", label="Mínimo en X (1.5 mil impresiones)")

# Rellenar la región factible
plt.fill_between(x1, x4_min, x4_limit, where=(x4_limit >= x4_min), color="grey", alpha=0.3)

# Configuración de la gráfica
plt.xlim(0, 5)
plt.ylim(0, 5)
plt.xlabel("Impresiones en Instagram (X1, miles)")
plt.ylabel("Impresiones en X (X4, miles)")
plt.title("Región Factible para Maximizar el Alcance")
plt.legend()
plt.grid(True)

# Mostrar la gráfica
plt.show()
```

**Explicación**

* Restricciones: Se calculan como funciones lineales, resolviendo la desigualdad de presupuesto para obtener X4 en función de X1. Las demás variables están en su valor mínimo permitido (1.5).

* Región factible: Es el área donde todas las restriccipnes son válidas simultaneamente.

* Gráfica: Se trazan las restricciones y se resalta la región factible.

**Resultado**

1. Restricciones de presupuesto:
- Las líneas azule sy naranjas representan las combinaciones posibles de impresiones en Instagram (X1) y X (X4) que satisfacen la restricción presupuestaria cuando tiktok y facebook están en su nivel mínimo permitido (1.5 mil impresiones).

- Estas restricciiones reflejan como se distribuye el presupuesto limitado entre plataformas.

2. Restricción mínima para X4:
- La línea roja punteada muestra que X4 (impresiones en X) debe ser al menos 1.5 mil para garantiel mínimo de visitas en esa plataforma.

3. Región factible:
- El área sombreada gris representa las combinciones de impresiones en Instagram (X1) y (X4) que satisfacen todas las restricciones simultáneamente.

- Solo dentro de esta región es posible asignar impresiones a las plataformas sin superar el presupuesto ni incumplir las restricciones mínimas de cada variable.

4. Interpretación:

- Si el valor de X1 (Impresiones en Instagram) aumenta, se reduce la cantidad disponible para X4 debido al costo relativamente alto de las impresiones en instagram ($222.71 / 1000).

- La región factible indica que aumentar el presupuesto total permitiría mayor flexibilidad en la asignación, expandiendo la región factible.

5. Decisión óptima:

- La solución óptima (encontrada en el modelo líneal) estará en algún vértice de la región factible, maximizando el alcance total de impresiones para todas las plataformas sin exceder el presupuesto.

#**¿A cuantas personas llegariamos?**

Podemos calcularlo con la relación entre las personas alcanzadas  con las impresiones.

La función objetivo es: **Z = 1000 * (x1 + x2 + x3 + x4)**

Esto implica que el número total de personas alcanzadas es directamente proporcional a la suma de las variables (x1 + x2 + x3 + x4), cada una representando miles de impresiones.

**Ejemplo de Objetivo:**

Para un alcance de 6000 personas, la suma total de las variables óptimas (x1 + x2 + x3 + x4) seria:

**x1 + x2 + x3 + x4 = 6000 / 1000 = 6 **

Esto significa que la solución óptima asignaria un total de 6 mil impresiones distribuidas entre las plataformas.

La distribución exacta de impresiones dependerá de los costos de cada plataforma y la restricción presupuestaria. Una posible distribución podría ser:

- x1 = 1.5 (mínimo para intagram).
- X2 = 1.5 (Mínimo para Tiktok).
- X3 = 1.5 (Minímo para Facebook).
- X4 = 1.5 (Minímo para X).

En este caso:

**x1 + x2 + x3 + x4 = 1.5 + 1.5 + 1.5 + 1.5 = 6 mil impresiones.**

Y el alcance sería:

**Z = 1000 * (6) = 6000 personas.**


**El modelo tiene como objetivo asignar las cantidades mínimas necesarias para cada plataforma y ajustarlas para maximizar el alcance, manteniendo el presupuesto total.**

**Solución óptima con PuLP**

```python
from pulp import LpProblem, LpVariable, LpMaximize, value

# Costos de cada red social por 1000 impresiones
cost_instagram = 222.71  # Instagram
cost_tiktok = 206.98     # TikTok
cost_facebook = 202.43   # Facebook
cost_x = 200             # X (estimado)

# Crear el modelo de programación lineal
model = LpProblem("Maximizar_Alcance", LpMaximize)

# Variables de decisión (en miles de impresiones)
x1 = LpVariable("Impresiones_Instagram", lowBound=1.5, cat="Continuous")
x2 = LpVariable("Impresiones_TikTok", lowBound=1.5, cat="Continuous")
x3 = LpVariable("Impresiones_Facebook", lowBound=1.5, cat="Continuous")
x4 = LpVariable("Impresiones_X", lowBound=1.5, cat="Continuous")

# Función objetivo: Maximizar el alcance total
#Se maximiza el total de impresiones (y por ende el alcance) en las plataformas de redes sociales.
model += (1000 * x1 + 1000 * x2 + 1000 * x3 + 1000 * x4), "Alcance_Total"

# Restricción de presupuesto
#Aseguramos que el presupuesto total no supere $1248.19 MXN, que es el costo mínimo para alcanzar las 1500 visitas por plataforma.
model += (cost_instagram * x1 + cost_tiktok * x2 +
          cost_facebook * x3 + cost_x * x4) <= 1248.19, "Presupuesto_Total"

# Resolver el modelo
model.solve()

# Mostrar resultados
print("Solución óptima: \n")
print(f"Impresiones en Instagram: {x1.varValue:.2f} mil \n")
print(f"Impresiones en TikTok: {x2.varValue:.2f} mil \n")
print(f"Impresiones en Facebook: {x3.varValue:.2f} mil \n")
print(f"Impresiones en X: {x4.varValue:.2f} mil \n")
print(f"Personas alcanzadas totales: {value(model.objective):.2f} \n")
print(f"Costo total: ${cost_instagram * x1.varValue + cost_tiktok * x2.varValue + cost_facebook * x3.varValue + cost_x * x4.varValue:.2f} MXN")
```

#**Conclusiones Generales de los Resultados del Modelo de Programación Lineal**

**1. Optimización del presupuesto:**

- El modelo utiliza eficientemente el presupuesto de $1248.19 MXN, asegurando que cada plataforma reciba el número mínimo de impresiones necesarias para alcanzar 1500 visitas.

- Esto resulta en un alcance total de 6000 personas distribuidas equitativamente entre las cuatro plataformas.

**2. Restricciones dominantes:**

- La principal limitación del modelo es el presupuesto disponible. Debido al costo relativamente alto de las impresiones en las plataformas, no es posible asignar impresiones adicionales para aumentar el alcance sin exceder el presupuesto.

- Las restricciones mínimas de 1.5 mil impresiones por plataforma también son cruciales, ya que garantizan que cada plataforma cumpla con los mínimos requeridos.

**3. Distribución equitativa:**

- La solución asigna las impresiones de forma equitativa entre las plataformas, lo que asegura un alcance balanceado. Cada plataforma contribuye exactamente con 1500 visitas, lo que indica que no se priorizó ninguna plataforma en particular.

**4. Limitaciones del alcance máximo:**

- Dado el presupuesto y los costos por mil impresiones (CPM), el alcance máximo alcanzable es limitado. El modelo sugiere que para aumentar el alcance total, sería necesario:
  - Aumentar el presupuesto disponible.

  - Priorizar plataformas más económicas como Facebook ($202.43) y X ($200), sacrificando impresiones en plataformas más costosas como Instagram ($222.71).

**5. Impacto práctico:**

- Este modelo asegura una campaña publicitaria eficiente con los recursos disponibles, cumpliendo con los mínimos requeridos en todas las plataformas.

- Sin embargo, no permite explorar estrategias de maximización del alcance total más allá de los mínimos debido a las restricciones de costo.
