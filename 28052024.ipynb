Si el dataset al inicio tiene una columna llamada ‘Columna1’,
  Borrarla del dataframe

```python
df_NBA=df_NBA.drop('Columna1', axis=1)
df_NBA.head()
```

Agrega otra variable al final del DataFrame, esta se llamará
Height_cms y será una conversión de la variable Height a centímetros

Con el punto 6, agregar una nueva columna si la estatura de
 Height_cms es igual o mayor a 2 metros, y se etiquetará como
‘Muy alto’

```python
df_NBA[['pies', 'pulgadas']]= df_NBA['Height'].str.split('-', expand=True)
df_NBA
```

```python
df_NBA['pulgadas'].replace(np.inf, df_NBA['pulgadas'][df_NBA['pulgadas'] != np.inf].max(), inplace=True)
df_NBA['pulgadas'] = df_NBA['pulgadas'].astype(str)
df_NBA
```

```python
df_NBA['pulgadas']= df_NBA['pulgadas'].map({'feb':2, 'ene':1, 'mar':3, 'abr':4, 'may':5, 'jun':6, 'jul':7, 'ago':8, 'sep':9, 'oct':10,'nov':11, 'dic':12}).astype(str)
df_NBA
```

```python
df_NBA['pulgadas'].fillna('0', inplace=True)
df_NBA
```

```python
df_NBA['pies'] = pd.to_numeric(df_NBA['pies'], errors='coerce').fillna(0).astype(int)
df_NBA
```

```python
df_NBA['pulgadas'] = pd.to_numeric(df_NBA['pulgadas'], errors='coerce').fillna(0).astype(int)
df_NBA
```

```python
df_NBA['Estatura_cm'] = df_NBA['pies'] * 30.48 + df_NBA['pulgadas'] * 2.54
df_NBA
```

```python
df_NBA.dtypes
```

8 con el paso número 8 asignar una variable que cuente cuántos
   jugadores se les colocó esta de etiqueta de ‘Muy alto’

```python
df_NBA['Name'] = df_NBA['Estatura_cm'].apply(lambda x: 'Muy alto' if x >= 200 else 'Bajo')
df_NBA
```
